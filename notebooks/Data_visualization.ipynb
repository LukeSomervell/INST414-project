```python
import pandas as pd

# Load cleaned dataset
file_path = "../data/processed/CEAS_08_cleaned.csv"
df = pd.read_csv(file_path)
```

```python

import matplotlib.pyplot as plt


###Visualization 1
# Set up figure size
plt.figure(figsize=(6,4))

# Plot bar chart of label counts
df['label'].value_counts().sort_index().plot(kind='bar', color=['green', 'red'], edgecolor='black')

# Add labels and title
plt.xticks([0, 1], ['Legitimate (0)', 'Phishing (1)'], rotation=0)
plt.xlabel('Email Class')
plt.ylabel('Count')
plt.title('Distribution of Legitimate vs Phishing Emails')
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
###Visualization 2
# Create a new column for body text length
df['body_length'] = df['body'].apply(len)

# Set up the figure
plt.figure(figsize=(8,5))

# Plot histograms of body length by label
for label, color, name in zip([0, 1], ['green', 'red'], ['Legitimate', 'Phishing']):
    subset = df[df['label'] == label]
    plt.hist(subset['body_length'], bins=50, alpha=0.6, label=name, color=color, edgecolor='black')

# Add labels and title
plt.xlabel('Body Text Length (characters)')
plt.ylabel('Number of Emails')
plt.title('Distribution of Email Body Length by Class')
plt.legend()
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
###Visualization 3
import seaborn as sns
import matplotlib.pyplot as plt

# Set up figure
plt.figure(figsize=(8,5))

# Density plot (KDE) for body lengths
sns.kdeplot(df[df['label'] == 0]['body_length'], label='Legitimate (0)', color='green', fill=True, alpha=0.5)
sns.kdeplot(df[df['label'] == 1]['body_length'], label='Phishing (1)', color='red', fill=True, alpha=0.5)

# Add labels and title
plt.xlabel('Body Text Length (characters)')
plt.ylabel('Density')
plt.title('Density Plot of Email Body Length by Class')
plt.legend()
plt.xlim(0, 5000)  # Optional: zoom in to see the important part better
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
###Visualization 4

# Create a new column for subject text length
df['subject_length'] = df['subject'].apply(len)

# Set up the figure
plt.figure(figsize=(8,5))

# Plot histograms of subject length by label
for label, color, name in zip([0, 1], ['green', 'red'], ['Legitimate', 'Phishing']):
    subset = df[df['label'] == label]
    plt.hist(subset['subject_length'], bins=50, alpha=0.6, label=name, color=color, edgecolor='black')

# Add labels and title
plt.xlabel('Subject Line Length (characters)')
plt.ylabel('Number of Emails')
plt.title('Distribution of Email Subject Length by Class')
plt.legend()
plt.grid(axis='y')

# Show the plot
plt.show()


```

```python
###Visualization 5

# Set up figure
plt.figure(figsize=(6,4))

# Plot bar chart showing URL presence grouped by label
df.groupby('label')['urls'].mean().plot(kind='bar', color=['green', 'red'], edgecolor='black')

# Add labels and title
plt.xticks([0, 1], ['Legitimate (0)', 'Phishing (1)'], rotation=0)
plt.ylabel('Average URL Presence (0 = None, 1 = At Least One)')
plt.xlabel('Email Class')
plt.title('Average URL Presence in Legitimate vs Phishing Emails')
plt.grid(axis='y')

# Show the plot
plt.show()
```
