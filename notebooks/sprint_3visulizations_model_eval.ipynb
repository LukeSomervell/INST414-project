```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn for improved style
sns.set(style="whitegrid")

# Model names and scores
models = ['Logistic Regression', 'Decision Tree', 'BERT']
accuracy = [0.72, 0.921, 0.99]
f1_scores = [0.763, 0.929, 0.99]

# Define color palette
colors = sns.color_palette("viridis", len(models))

# Accuracy Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=models, y=accuracy, palette=colors)
plt.ylim(0.6, 1.0)
plt.ylabel("Accuracy", fontsize=12)
plt.title("Model Accuracy Comparison", fontsize=14, weight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
for i, score in enumerate(accuracy):
    plt.text(i, score + 0.01, f"{score:.2f}", ha='center', fontsize=11, weight='bold')
plt.tight_layout()
plt.show()

# F1 Score Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=models, y=f1_scores, palette=colors)
plt.ylim(0.6, 1.0)
plt.ylabel("F1 Score (Phishing Class)", fontsize=12)
plt.title("Model F1 Score Comparison", fontsize=14, weight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
for i, score in enumerate(f1_scores):
    plt.text(i, score + 0.01, f"{score:.2f}", ha='center', fontsize=11, weight='bold')
plt.tight_layout()
plt.show()
```
